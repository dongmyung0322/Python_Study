```python
# 생성자 소멸자
class myClass:
  def __init__(self):  # 생성자 -> 클래스 생성시 작동됨
    print("class is created")

my=myClass()
del my
```

```python
class Person:
  def __init__(self,name,age):
    self.Name=name
    self.Age=age
  def __del__(self):  #소멸자 -> 소멸될때 작동됨
    print("{0}, {1}세 잘살다 갑니다".format(self.Name,self.Age))
  def printName(self):
    print(self.Name,self.Age)

p=Person("Park",30)
p.printName()
print(p.Name)
del p
```

```python
# 정적(static) 메소드
class CounterManager:
  InsCount=0
  def __init__(self):
    CounterManager.InsCount+=1
  def printInstanceCount():  #정적 메소드
    print("Instance count: ",CounterManager.InsCount)
  staticCount=staticmethod(printInstanceCount)


a=CounterManager()
b=CounterManager()
c=CounterManager()

CounterManager.printInstanceCount() # class를 직접 부르는것은 금기
#a.printInstanceCount()
a.staticCount()
```

```python
#심화 상속
class Person:
  def __init__(self,name,phone):
    self.Name=name
    self.Phone=phone
  def PrintInfo1(self):
    print("{0}: {1}".format(self.Name,self.Phone))

class Student(Person):
  pass
  def __init__(self,name,phone,score):
    Person.__init__(self,name,phone)
    #self.Name=name
    #self.Phone=phone
    self.Score=score
  def PrintInfo2(self):
    print("{0}: {1}, {2}점".format(self.Name,self.Phone,self.Score))
  def PrintInfo1(self):
    #return super().PrintInfo1()
    print("{0}의 점수는 {1}".format(self.Name,self.Score))

p=Person("Lee","010-1111-2222")
s=Student("Park","010-2222-3333",76)
print(p.__dict__)
print(s.__dict__)

p.PrintInfo1()
s.PrintInfo2()
s.PrintInfo1()
```

```python
# 상속 관계 확인 issubckass(a<)
print(issubclass(Student,Person))
print(issubclass(Person,Student))
```

```python
#다중상속
class Tiger:
  def Jump(self):
    print("호랑이 점프")

class Lion:
  def Bite(self):
    print("사자의 물기")

class Liger(Tiger,Lion):
  def Play(self):
    print("라이거 놀기")

a=Liger()
a.Jump()
a.Bite()
a.Play()
```
