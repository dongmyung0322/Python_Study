```python
# 8진수
#정수 앞에 0o를 붙이면 8진수
print(0o10)
print(type(0o10))
# 정수앞 0b는 2진수, 0x는 16진수
print(0b10,0x10)
```

```python
#10진수를 받아 각 진수로 변환하는법 (결과=문자형)
print(oct(38),type(oct(38))) #10진수 38 -> 8진수
print(hex(38),type(hex(38))) #10진수 38 -> 16진수
print(bin(38),type(bin(38))) #10진수 38 -> 2진수
```

```python
print(type(3.14))
print(type(314e-2))
print(type(3-4j)) # 복소수
x=3-4j
print(x.imag) #허수부
print(x.real) #실수부
print(x.conjugate()) # 켤레 복소수 확인
```
