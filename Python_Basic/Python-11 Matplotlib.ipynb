```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Style 지정
plt.style.use ("classic")
```

```python
import numpy as np
x=np.linspace(0,10,100) #0~10사이 값 100개 생성
print(np.sin(x))
graph=plt.plot(x,np.sin(x))
graph=plt.plot(x,np.cos(x))
print(graph)
```

```python
fig=plt.figure()
plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
fig.savefig("my_fig.png")
```

```python
#색 표기법
plt.plot(x,np.sin(x), color = "black")
plt.plot(x,np.sin(x-1), color = "y") # black= k
plt.plot(x,np.sin(x-2), color = "0.75") # 0(흰색)~1(검정색)
plt.plot(x,np.sin(x-3), color = "#FFDD44")
plt.plot(x,np.sin(x-4), color = (1.0,0.2,0.3)) # (R,G,B)
plt.plot(x,np.sin(x-5), color = "chartreuse")
```

```python
# 선의 스타일 
plt.plot(x, x, linestyle="solid")
plt.plot(x, x+1, linestyle="dashed")
plt.plot(x, x+2, linestyle="dashdot")
plt.plot(x, x+3, linestyle="dotted")

plt.plot(x, x+5, linestyle="-")
plt.plot(x, x+6, linestyle="--")
plt.plot(x, x+7, linestyle="-.")
plt.plot(x, x+8, linestyle=":")
```

```python
# 라벨 붙이기
plt.plot(x, np.sin(x))
plt.title("A sine curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), "-r",label = "sin(x)")
plt.plot(x, np.cos(x), ":b", label = "cos(x)")
plt.axis("equal")
plt.legend()
plt.title("Sine & Cosine")
plt.xlabel("x")
plt.ylabel("y")
```

```python
# 산점도의 표현 (scatter plot)
x=np.linspace(0,10,30)
y=np.sin(x)
plt.plot(x, y, "o", color="k" ) #marker
```

```python
rng=np.random.RandomState(0) # 시드 값을 기준으로 랜덤값을 섞지만 결과는 같게나옴
#rng.rand(5) 섞은 값중 랜덤으로 5개 뽑기
plt.style.use("bmh")

for marker in ["o", ",", ",", "x","+","v", "^", ">", "<", "s", "d"]:
  plt.plot(rng.rand(5), rng.rand(5), marker, label="marker={0}".format(marker)) 
  plt.legend(numpoints=1)
plt.axis([0,1.8,0,1])
```

```python
# plt.scatter을 사용한 산점도
plt.style.use("seaborn")
rng=np.random.RandomState(0)
x=rng.randn(100) #-1~1 n개
y=rng.rand(100) #0~1 n개
size=rng.rand(100)*1000
colors=rng.rand(100)
plt.scatter(x, y, s=size, alpha=0.3, c=colors, cmap="viridis")

```

```python
from sklearn.datasets import load_iris
iris=load_iris()
features=iris.data.T

for i in range (0,4):
  print(iris.feature_names[i])
print(iris.target)

plt.style.use("classic")
for i in range (0,3):
  plt.scatter(features[2],features[3], c=iris.target, s=features[3]*100, alpha=0.2, cmap="viridis", label="target={0}".format(iris.target_names[i]))
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
iris.target
```
