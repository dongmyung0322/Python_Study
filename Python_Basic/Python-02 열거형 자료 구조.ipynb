```python
# List
colors=["red","green","gold"]
colors.append("blue")
colors.insert(1,"black") #원하는 위치에 끼워넣기
print(colors)
colors.extend(["white","gray"]) #다른 리스트를 뒤에 붙이기
colors+=["red"]
colors+="red"
print(colors)
```

```python
print(colors.index("red"))
print(colors.index("red",1))
print(colors.count("red"))
```

```python
print(colors.pop())
print(colors.pop(1))
print(colors)
colors.remove("gold")
print(colors)
```

```python
colors.sort()
print(colors)
colors.reverse()
print(colors)
```

```python
# 세트(set) --> 합집합 교집합을 구할 수 있음
a={1,2,3}
b={3,4,5}
print(type(a))
```

```python
print(a.union(b)) #합집합
print(a.intersection(b)) #교집합
print(a-b) #차집합
print(a|b) #합집합
print(a&b) #교집합
```

```python
# in 연산자를 통한 특정값 유무 확인
a={1,2,3}
print(1 in a)
print("gold" in colors)
```

```python
# Dictionary 
d=dict(a=1,b=2,c=3)
print(d)
print(type(d))


color={"apple":"red", "banana":"yellow"}
print(color["apple"])
#print(color[1]) dict에는 index값 없음
#print(color["cherry"]) 없는값 부르면 에러

color["cherry"]="red"
color["apple"]="green"
print(color)

print(color.keys())
print(color.values())
print(color.items())
```

```python
#for문으로 데이터 가져오기
for i in color.keys():
  print(i)
  print(color[i])
for v in color.values():
  print(v)
for (key, value) in color.items():
  print(key, value)
```

```python
# Dict에서 불필요한 아이템의 삭제
del color["cherry"]
print(color)
color.clear()
print(color)
```

```python
# data 타입은 혼용가능
d={"age":40.5, "job":[1,2,3], "name":{"Kim":2},"Cho":1}
print(d)
for value in d.values():
  print(type(value))
```

```python
# 얕은 복사와 깊은 복사

a=[1,2,3]
b=a
print(b)
a[0]=38
print(a,b)

a=[1,2,3]
b=a[:] #call by value
a[0]=38
print(a,b)
```

```python
# copy module 사용하는법
import copy

a=[1,2,3]
b=copy.deepcopy(a)
a[0]=38
print(a,b)
```
