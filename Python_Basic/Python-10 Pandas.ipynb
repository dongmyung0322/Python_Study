```python
import pandas as pd
from pandas import Series,DataFrame
```

```python
#Series 객체
fruit = Series([2500,3800,1200,1600], index = ["apple","banana","cherry","orange"])
print(fruit)
print(fruit.values)
print(fruit.index)
```

```python
# Dictionary -> Series
dictionary = {"apple":2500, "banana":3800, "cherry":1200, "orange":1600}
fruit=Series(dictionary)

print(fruit)
```

```python
# 이름 지어보기
fruit.name = "fruitPrice"
fruit.index.name = "fruitName"
print(fruit)
```

```python
# DataFrame 객체
fruitData = ({"fruitName":["apple","banana","cherry","orange"],"fruitPrice":[2500,3800,1200,1600], "num":[100,200,150,300]})
print(fruitData)
fruitFrame = DataFrame(fruitData,columns=["fruitPrice", "num", "fruitName"])
print(fruitFrame)

#컬럼방식
print(fruitFrame["fruitName"])
#속성방식
print(fruitFrame.fruitName)
fruitFrame["year"] = 2023
print(fruitFrame)
```

```python
variable = Series([4,2,1], index=[0,2,3])
fruitFrame["stock"]=variable
print(fruitFrame)
```

```python
#데이터 구조에서 삭제
fruit = Series([2500,3800,1200,1600], index=["apple","banana","cherry","orange"])
new_fruit=fruit.drop("banana") # 실제 series에서 삭제되지는 않음
print(new_fruit)
print(fruit)
```

```python
fruitFrame = DataFrame(fruitData, index=fruitData["fruitName"], columns=["fruitPrice", "num"])
print(fruitFrame)
print(fruitFrame.drop(["apple","orange"]))
print(fruitFrame.drop(["num"],axis=1)) #열삭제
```

```python
#DataFrame 슬라이싱
print(fruitFrame["banana":"cherry"])
```

```python
#데이터의 기본 연산
fruit1 = Series([5,9,10,3], index=["apple", "banana", "cherry", "orange"])
fruit2 = Series([3,2,9,5,10], index=["apple", "orange", "banana", "kiwi", "cherry"])
print(fruit1+fruit2)

fruitData1 = {"Ohio":[4,8,3,5], "Texas":[0,1,2,3]}
fruitData2 = {"Ohio":[3,0,2,1,7], "Colorado":[5,4,3,6,0]}
fruitFrame1 = DataFrame(fruitData1, columns = ["Ohio","Texas"], index = [fruit1.index])
fruitFrame2 = DataFrame(fruitData2, columns = ["Ohio","Colorado"], index =[fruit2.index])
print(fruitFrame1)
print(fruitFrame2)
print(fruitFrame1+fruitFrame2)
```

```python
#데이터의 정렬
print(fruit)
print(fruit.sort_values(ascending = False))
```

```python
#DataFrame 객체의 정렬
print(fruitFrame)
print(fruitFrame.sort_index(ascending = False))
print(fruitFrame.sort_values(by=["fruitPrice", "num"], ascending=[True,False]))
```

```python
#기초 분석(기술 통계량) 
#count NA를 제외한 개수
#min, max 최소 최댓값
#comprod 누적합
#sum 합
#mean 평균
#median 중앙값
#quantile 분위수
#var 표본분산
#std 표본 정규분산(표준편차)
#describe 요약 통계량

german_sample = pd.read_csv("http://freakonometrics.free.fr/german_credit.csv")
```

```python
german_sample
```

```python
print(german_sample.columns.values)
print(list(german_sample.columns.values))
```

```python
german = german_sample[["Creditability", "Duration of Credit (month)", "Purpose", "Credit Amount"]]
german
```

```python
print(german.min())
print(german.max())
print(german.mean())
print(german.describe())
```

```python
#상관관계
german = german_sample[["Duration of Credit (month)", "Credit Amount", "Age (years)"]]
german.head(5)
```

```python
german.corr()
```

```python
#핵심기능 Group By
german=german_sample[["Credit Amount", "Type of apartment"]]
german.head()
```

```python
german_grouped = german.groupby(german["Type of apartment"])
print(german_grouped.mean())
```

```python
german=german_sample[["Credit Amount", "Type of apartment","Purpose"]]
german_grouped=german.groupby([german["Purpose"],german["Type of apartment"]])
print(german_grouped.mean())
```

```python
german = german_sample[["Type of apartment", "Sex & Marital Status", "Credit Amount"]]
german.head()
for type, g in german.groupby(german["Type of apartment"]):
  print(g.head())
```

```python
births=pd.read_csv("https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv")
births.head()
```

```python
births["decades"] = births["year"]//10*10 #연대 구하기
```

```python
births.pivot_table("births",index="decades",columns="gender",aggfunc="sum")
```

```python
births.pivot_table("births",index="decades",columns="gender",aggfunc="sum").plot()
```

```python
births.pivot_table("births",index="year",columns="gender",aggfunc="sum").plot()
```
