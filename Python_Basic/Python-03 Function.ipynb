```python
#함수의 인자 (Parameter)
def times(a=10,b=20):
  return a*b
print(times())
print(times(1,2))
print(times(100))
print(times(b=2))
```

```python
def connectURL(protocol="https",server="naver.com",port="80"):
  address=protocol+"://"+server+":"+port
  return address

print(connectURL("http","naver.com","80"))
```
