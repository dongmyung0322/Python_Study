```python
from azure.storage.file import FileService
file_service = FileService(account_name="labuser109storage",account_key="
```

```python
# 공유 파일 생성
file_service.create_share('myshare')
```

```python
# directory 생성
file_service.create_directory("myshare","sampledir")
```

```python
# 파일 업로드
from azure.storage.file import ContentSettings
file_service.create_file_from_path(
    "myshare",None ,"myfile", "IU.jpg"  #업로드할 파일위치, 업로드할 directory 위치 (젤 상단이면 none) ,업로드 된 후의 파일이름, 업로드할 파일이름
)
```

```python
# directory 내용의 확인
generator = file_service.list_directories_and_files("myshare/sampledir")
for file_or_dir in generator:
  print(file_or_dir.name)
```

```python
# 파일을 가져오는 코드
file_service.get_file_to_path("myshare", None, "myfile", "IU-DOWNLOAD.jpg")
#파일 위치, 다운받을 경로(상관없으면 None), 다운받을 파일이름, 다운받은후 파일이름 
```

```python
# snapshot 
metadata = {"foo":"bar"}
snapshot=file_service.snapshot_share("myshare", metadata=metadata)
```

```python
share = list(file_service.list_shares(include_snapshots=True)) #snapshot이 있는 share파일들의 리스트를 불러오기
directories_and_files = list(file_service.list_directories_and_files("myshare",snapshot="2023-04-20T05:48:33.0000000Z"))
for file_or_dir in directories_and_files:
  print(file_or_dir.name)
```

```python
#스냅샷 삭제
file_service.delete_share("myshare", snapshot="2023-04-20T05:48:33.0000000Z")
```

```python
file_service.delete_share("myshare",None,"myfile") #myshare의 최상위 위치의(None) myfile삭제
```
